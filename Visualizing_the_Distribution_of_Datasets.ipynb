# Visualizing the Distribution of Dataset


---


1.   Univariate Distribution


2.   Bivariate Distribution




```
import numpy as np
import pandas as pd
import seaborn as sns
from scipy import stats
```

```
sns.set(color_codes = True)
```

## Plotting Univariate Distributions

```
x = np.random.normal(size=100)
sns.distplot(x)
```

```
# removing the kernel density and adding rug
sns.distplot(x,kde = False,rug=True)
```

```
# varying bin count
sns.distplot(x,kde = False,rug=True,bins = 20)
```

```
# Plotting kde
sns.distplot(x,hist = False,rug=True)
```

```
# varying bandwidth
sns.kdeplot(x)
sns.kdeplot(x,bw=0.2,label = "bw=0.2")
sns.kdeplot(x,bw=2,label="bw=2")
```

```
# Fitting parametric distributions
x = np.random.gamma(6,size=200)
sns.distplot(x,kde=False,fit=stats.gamma)
```

## Plotting Bivariate Distribution

```
mean, cov = [0, 1], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean, cov, 200)
df = pd.DataFrame(data, columns=["x", "y"])
```

```
# scatterplot
sns.jointplot("x","y",data=df)
```

```
# hexbin plots
sns.jointplot("x","y",kind="hex",data=df)
```

```
# kde
sns.jointplot("x","y",kind="kde",data=df)
```

## Visualizing Pairwise Relationships

```
iris = sns.load_dataset("iris")
sns.pairplot(iris)
```

```
# adding hue semantic
sns.pairplot(iris,hue="species")
```
