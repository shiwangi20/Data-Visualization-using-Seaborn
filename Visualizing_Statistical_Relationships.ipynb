# Visualizing Statistical Relationships using Seaborn

## Author : Shiwangi Garg


Dataset Analyzed : tips, fmri, dots

Semantics used : hue, style, size

```
import numpy as np
import seaborn as sns
```

```
sns.set(style='darkgrid')
tips = sns.load_dataset('tips')
```

# Relating variables with Scatter Plots

```
# when data is numeric generally, scatterplot() is used
# scatter is the default kind of replot()
sns.relplot(x='total_bill',y='tip',data=tips)
```

```
# using hue semantic
sns.relplot(x='total_bill',y='tip',hue='smoker',data=tips)
```

```
# using style semantic
sns.relplot(x='total_bill',y='tip',hue='smoker',style='smoker',data=tips)
```

```
# using size semantic
sns.relplot(x='total_bill',y='tip',size = 'size',sizes = (15,200),data=tips)
```

```
# using quantitative hue
sns.relplot(x='total_bill',y='tip',hue='size',data=tips)
```

# Aggregation and representing Uncertainity

```
fmri = sns.load_dataset('fmri')
fmri.head()
```

```
# lineplot() is used to represent the variation of data
sns.relplot(x='timepoint',y='signal',kind='line',data=fmri)
```

```
# removing the confidence intervals
sns.relplot(x='timepoint',y='signal',kind='line',ci=None,data=fmri)
```

```
# turning off estimator
sns.relplot(x='timepoint',y='signal',kind='line',estimator = None,data=fmri)
```

```
# adding semantic hue
sns.relplot(x='timepoint',y='signal',kind='line',hue = 'event',data=fmri)
```

```
# adding semantic style
sns.relplot(x='timepoint',y='signal',kind='line',style='region',data=fmri)
```

```
dots = sns.load_dataset('dots').query("align == 'dots'")
dots.head()
```

```
sns.relplot(x='time',y='firing_rate',kind='line',hue='coherence',style='choice',size='choice',data=dots)
```

# Showing multiple replationships with facets

```
sns.relplot(x='total_bill',y='tip',col='time',hue='smoker',data=tips)
```
